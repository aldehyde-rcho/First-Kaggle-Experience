```python
df=pd.read_csv("/kaggle/input/glass/glass.csv")
```

```python
df.head()
df.info()
df.shape
df.describe()
```

```python
#categorizing into features and lables
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X.head()
y.head()
y.unique()
y.value_counts()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25)
```

```python
from sklearn import tree
#building decision tree model
dtc=tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
sn=[]
score=[]
for i in range(1,100):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25)
    dtc=tree.DecisionTreeClassifier(random_state=0)
    dtc.fit(X_train, y_train)
    sn.append(i)
    score.append(dtc.score(X_test,y_test))

    result=pd.DataFrame()
    result["SN"]=sn
    result["Score"]=score
```

```python
result.describe()
```

```python
result
```

```python
result.to_csv("r1.csv")
```

```python
#normalizing data
from  sklearn import preprocessing
minmax_scale=preprocessing.MinMaxScaler(feature_range=(1,10))
X_minmax=minmax_scale.fit_transform(X)
```

```python
X_minmax
```

```python
X1=pd.DataFrame(columns=X.columns,data=X_minmax)
X1
```

```python
sn=[]
score=[]
for i in range(1,100):
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X1, y,  stratify=y,  test_size=0.25)
    dtc = tree.DecisionTreeClassifier(random_state=0)
    dtc.fit(X_train, y_train)
    sn.append(i)
    score.append(dtc.score(X_test,y_test))

result=pd.DataFrame()
result["SN"]=sn
result["Score"]=score
print(result)    
```
